### 1- Cuáles son las condiciones para que un sistema sea considerado como lineal?

Un sistema puede ser considerado como lineal si cumple con la propiedad de súper posición y homogeneidad, es decir si cumple con lo siguiente.

1. Superposición: La respuesta total a una combinación lineal de entradas es igual a la suma de las respuestas individuales a cada entrada por separado.
$$f(x + y) = f(x) + f(y)$$

2. Homogeneidad: La respuesta a una entrada escalada es igual a la entrada escalada de la respuesta.
$$f(ax) = af(x)$$

### 2- Por qué es conveniente modelar los problemas con sistemas lineales?

La modelización de problemas con sistemas lineales es conveniente por su simplicidad, generalidad, eficiencia computacional y la posibilidad de tener una interpretación física o intuitiva de las soluciones.

### 3- Qué es un tensor y cómo se relaciona con los vectores?

Un tensor es un objeto matemático que generaliza el concepto de vector en varias dimensiones. Se puede representar por un conjunto de números que se organizan en una estructura multilinear, es decir, una estructura que es lineal en cada una de sus variables.
Los vectores son un caso especial de tensores, específicamente, son tensores de rango 1, lo que significa que son estructuras lineales que tienen una sola entrada o índice.

### 4- Cuál es la utilidad de la norma de Frobenius en Machine Learning?

La norma de Frobenius se utiliza como una penalización en técnicas de regularización como la regresión ridge. La inclusión de la norma de Frobenius en la función de pérdida ayuda a controlar el sobreajuste y a mejorar la generalización del modelo.

También la norma de Frobenius se utiliza en el cálculo de la descomposición en valores singulares (SVD) de una matriz.

Por otra parte la norma de Frobenius se utiliza como una medida de distancia entre dos matrices en algoritmos de optimización como el descenso de gradiente estocástico (SGD). 

###  5- Cómo se puede interpretar una matriz en términos de vectores, refiérase a la linealidad?

Una matriz puede interpretarse en términos de vectores como una transformación lineal que mapea un vector de entrada en otro vector de salida, y la linealidad de la matriz se refleja en su aditividad y homogeneidad. Estas propiedades son fundamentales en la teoría de matrices y en aplicaciones prácticas de las mismas.

### 1- Demuestre analíticamente si los siguientes sistemas son lineales o no lineales (Muestre todos los pasos). Considere a u(t) como la entrada del sistema, mientras que g(t) es la salida.

Para demostrar si una función es lineal o no, debemos verificar si cumple con las dos propiedades de la linealidad:

La propiedad de superposición: 
$$g(t_1 + t_2) = g(t_1) + g(t_2)$$

La propiedad de homogeneidad: 
$$g(\alpha t) = \alpha g(t)$$


### a) $g(t) = u(t)^2$

Propiedad de superposición:

$$g_1(t) = u_1(t)^2$$

$$g_2(t) = u_2(t)^2$$

$$g(t) = u(t)^2 = (u_1(t) + u_2(t))^2$$

$$g(t) = u_1(t)^2 + 2u_1(t)u_2(t) + u_2(t)^2$$

$$g(t) = g_1(t) + g_2(t) = u_1(t)^2 + u_2(t)^2$$

$$u_1(t)^2 + u_2(t)^2 \neq u_1(t)^2 + 2u_1(t)u_2(t) + u_2(t)^2$$

Propiedad de homogeneidad:

$$g(t) = u(t)^2$$

$$g(at) = (u(at))^2 = u(at)^2$$

$$g(at) = u(at)^2 \neq ag(t) $$

Por lo tanto, $g(t) = u(t)^2 \textbf{ no es lineal.}$

### b) $g(t) = u(t^2)$

Propiedad de superposición:

$$g_1(t) = u_1(t^2)$$

$$g_2(t) = u_2(t^2)$$

$$g(t) = u(t^2) = u_1(t^2) + u_2(t^2)$$

$$g(t) = g_1(t) + g_2(t) = u_1(t^2) + u_2(t^2)$$

$$u_1(t^2) + u_2(t^2) = u_1(t^2) + u_2(t^2)$$

Propiedad de homogeneidad:

$$g(t) = u(t^2)$$

$$g(at) = u((at)^2) = u(a^2t^2) = au(t^2)$$

$$ag(t) = au(t^2) = g(at)$$

Por lo tanto, $g(t) = u(t^2) \textbf{ es lineal.}$

### c) $g(t) = u(t)sin(5t)$

Propiedad de superposición:

$$g_1(t) = u_1(t)sin(5t)$$

$$g_2(t) = u_2(t)sin(5t)$$

$$g(t) = (u_1(t) + u_2(t))sin(5t)$$

$$g(t) = u_1(t)sin(5t) + u_2(t)sin(5t) = g_1(t) + g_2(t)$$

Propiedad de homogeneidad:

$$g(t) = u(t)sin(5t)$$

$$g(at) = (au(t))sin(5at)$$

$$g(at) = au(t)sin(5t) = ag(t)$$

Por lo tanto, $g(t) = u(t)sin(5t) \textbf{ es lineal.}$

### d) $g(t) = 5u(t)+6$

Propiedad de superposición:

$$g_1(t) = 5u_1(t)+6$$

$$g_2(t) = 5u_2(t)+6$$

$$g(t) = 5u(t) + 6 = 5(au_1(t) + u_2(t))+6$$

$$g(t) = 5a_1(t) + 5a_2(t) + 6

$$g_1(t) + g_2(t) = 5u_1(t) + 5u_2(t) + 12 \neq g(t)

Propiedad de homogeneidad:

$$g(t) = 5u(t)+6$$

$$g(at) = 5(u(at))+6$$

$$ag(t) = 5au(t)+6 \neq g(at)$$

Por lo tanto, $g(t) = 5u(t)+6 \textbf{ no es lineal.}$

### e) $g(t) = \int_{-\infty}^{\infty} u(t)$

Propiedad de superposición:

$$g_1(t) = \int_{-\infty}^{\infty} u_1(t)$$ 

$$g_2(t) = \int_{-\infty}^{\infty} u_2(t)$$

$$g(t) = \int_{-\infty}^{\infty} u(t) = \int_{-\infty}^{\infty} (u_1(t) + u_2(t))dt$$

$$g(t) = \int_{-\infty}^{\infty} u_1(t)dt + \int_{-\infty}^{\infty} u_2(t)dt = g_1(t) + g_2(t)$$


Propiedad de homogeneidad:

$$g(t) = \int_{-\infty}^{\infty} u(t)$$

$$g(t) = \int_{-\infty}^{\infty} u(t)dt$$

$$g(kt) = \int_{-\infty}^{\infty} u(kt)dt$$

$$g(kt) = \frac{1}{k}\int_{-\infty}^{\infty} u(s)ds = \frac{1}{k} g(t)$$

Por lo tanto, $g(t) = \int_{-\infty}^{\infty} u(t) \textbf{ es lineal}.$

## Tercer parte - VECTORES

### 1- Demuestre que si dos vectores $\vec{x}$, $\vec{y}$ son paralelos, entonces se cumple $ \vec{x} = k\vec{y}, k ∈ R $

Para demostrar esto, podemos usar la definición de vectores paralelos, que establece que dos vectores son paralelos si tienen la misma dirección o son opuestos en dirección.

Si $\vec{x}$ y $\vec{y}$ tienen la misma dirección, entonces existe una constante $k$ positiva tal que $\vec{x} = k\vec{y}$, ya que ambos vectores apuntan en la misma dirección y la magnitud de $\vec{x}$ es $k$ veces la magnitud de $\vec{y}$.

Si $\vec{x}$ y $\vec{y}$ son opuestos en dirección, entonces existe una constante $k$ negativa tal que $\vec{x} = k\vec{y}$, ya que $\vec{x}$ y $\vec{y}$ apuntan en direcciones opuestas y la magnitud de $\vec{x}$ es igual a la magnitud de $\vec{y}$ multiplicada por un factor negativo.

Por lo tanto, en ambos casos, se cumple que $\vec{x} = k\vec{y}$ para alguna constante $k \in \mathbb{R}$, lo que demuestra que dos vectores paralelos pueden expresarse en términos de un múltiplo escalar.

### 2- Dado los vectores $\vec{x} = \left( \begin{array}{c} -9 \\ 3 \\ 5 \end{array} \right)$ , $\vec{y} = \left( \begin{array}{c} 3 \\ 0 \\ -11 \end{array} \right)$, encuentre de manera analítica el ángulo de separación, distancia $l_{2}$ y el vector unitario para cada uno de ellos.

El ángulo de separación entre los vectores $\vec{x}$ y $\vec{y}$, primero se debe calcular su producto punto y magnitudes:

$$ \vec{x} \cdot \vec{y} = (-9)(3) + (3)(0) + (5)(-11) = -27 - 55 = -82 $$


$$ \vec{x} = \sqrt{(-9)^2 + 3^2 + 5^2} = \sqrt{115} $$ 


$$ \vec{y} = \sqrt{3^2 + 0^2 + (-11)^2} = \sqrt{130} $$

Por lo tanto, el coseno del ángulo de separación es:

$$\cos \theta = \frac{\vec{x} \cdot \vec{y}}{\left| \vec{x} \right| \left| \vec{y} \right|} = \frac{-82}{\sqrt{115} \sqrt{130}} \approx -0.6758$$

$$\theta \approx \cos^{-1}(-0.6758) \approx 129.2^\circ$$

---

Para encontrar la distancia $l_{2}$ entre los vectores $\vec{x}$ y $\vec{y}$, primero encontramos la diferencia entre los dos vectores:

$$ \vec{x} - \vec{y} = \left( \begin{array}{c} -9 \\ 3 \\ 5 \end{array} \right) - \left( \begin{array}{c} 3 \\ 0 \\ -11 \end{array} \right) = \left( \begin{array}{c} -12 \\ 3 \\ 16 \end{array} \right) $$

Luego, calculamos la norma $l_{2}$ de este vector diferencia:

$$ ||\vec{x} - \vec{y}||_{2} = \sqrt{(-12)^{2} + 3^{2} + 16^{2}} = \sqrt{289} = 17 $$

---

Por último dividimos cada vector por su norma para obtener el vector unitario correspondiente:

$$\hat{x} = \frac{\vec{x}}{|\vec{x}|} = \frac{1}{\sqrt{115}} \left( \begin{array}{c} -9 \\ 3 \\ 5 \end{array} \right) = \left( \begin{array}{c} -\frac{9}{\sqrt{115}} \\ \frac{3}{\sqrt{115}} \\ \frac{5}{\sqrt{115}} \end{array} \right)$$

$$\hat{y} = \frac{\vec{y}}{|\vec{y}|} = \frac{1}{\sqrt{130}} \left( \begin{array}{c} 3 \\ 0 \\ -11 \end{array} \right) = \left( \begin{array}{c} \frac{3}{\sqrt{130}} \\ 0 \\ -\frac{11}{\sqrt{130}} \end{array} \right)$$

$$z = \frac{155}{9} + \frac{11x}{3} - \frac{92y}{9}$$

$$\frac{\partial z}{\partial x} = \frac{11}{3} \quad \text{,} \quad \frac{\partial z}{\partial y} = -\frac{92}{9} \quad \text{,} \quad \frac{\partial z}{\partial z} = 0$$

Entonces, el vector gradiente del plano es:

$$\nabla f = \frac{11}{3} \hat{i} - \frac{92}{9} \hat{j} + 0 \hat{k}$$

## Quinta parte - MATRICES

### 1- Dada la matriz $ A = \left( \begin{matrix} 1 & 4 \\ -4 & -7 \\ \end{matrix} \right) $ calcule los eigen vectores y eigen valores de manera analítica, posteriormente utilice la función eig(A) para obtener la matriz de los autovalores y los auto vectores

Para calcular los eigen valores y eigen vectores de la matriz $A$, necesitamos resolver la ecuación característica:

$$\det(A - \lambda I) = 0$$

En este caso, la ecuación característica es:

$$\det \left( \begin{matrix} 1 - \lambda & 4 \\ -4 & -7 - \lambda \ \end{matrix} \right) = 0$$

Expandiendo esta determinante, obtenemos:

$$(1 - \lambda)(-7 - \lambda) - 4(-4) = \lambda^2 + 6\lambda + 9 = 0 $$ 
$$ (\lambda + 3)^2 = 0$$

$$\lambda_1 = -3 \text{, } \lambda_2 = -3$$

Para encontrar los eigen vectores correspondientes, debemos resolver el sistema de ecuaciones $(A - \lambda I) \mathbf{v} = \mathbf{0}$ para cada eigen valor. 

Para $\lambda_1 = \lambda_2 = -3$, esto es:

$$\left( \begin{matrix} 1 - (-3) & 4 \\ -4 & -7 - (-3) \end{matrix} \right) \left( \begin{matrix} x \\ y \end{matrix} \right) = \left( \begin{matrix} 4 & 4 \\ -4 & -4 \end{matrix} \right) \left( \begin{matrix} x \\ y \end{matrix} \right) = \left( \begin{matrix} 0 \\ 0 \end{matrix} \right)$$

$$\begin{align*} 4x + 4y &= 0 \\ -4x - 4y &= 0 \end{align*}$$

Los eigen vectores correspondientes a $\lambda_1 = \lambda_2 = -3$ son de la forma $\mathbf{v}_1 = \begin{pmatrix} -1 \\ 1 \end{pmatrix}$.

```python
import numpy as np
  
  
A = np.mat("1 4;-4 -7")


evalues, evectores = np.linalg.eig(A)

print(A)
print("")
print(evalues)
print("")
print(evectores)
```
